```python
import sqlite3

# Connect to SQLite database (or create it if it doesn't exist)
conn = sqlite3.connect('atm_machine.db')

# Create a cursor object
cursor = conn.cursor()

# Create table for users if it doesn't exist
cursor.execute('''
    CREATE TABLE IF NOT EXISTS users (
        id INTEGER PRIMARY KEY,
        balance INTEGER NOT NULL,
        pin INTEGER NOT NULL
    )
''')

# Initialize the database with a default user (optional)
cursor.execute('SELECT COUNT(*) FROM users')
if cursor.fetchone()[0] == 0:
    cursor.execute('INSERT INTO users (balance, pin) VALUES (?, ?)', (5000, 1234))
    conn.commit()

# Get the user's balance and pin from the database
cursor.execute('SELECT balance, pin FROM users WHERE id = ?', (1,))
user = cursor.fetchone()
bal = user[0]
pin = user[1]

while True:
    print(""" 0.Exit \n 1. Show balance \n 2. Withdraw cash \n 3. Deposit cash \n 4. Change pin""")
    a = int(input("Enter your choice: "))
    
    if a == 0:
        break
    elif a in (1, 2, 3, 4):
        trials = 3
        while trials != 0:
            enter_pin = int(input("Enter your pin: "))
            if enter_pin == pin:
                if a == 1:
                    print(f"Your balance is: {bal}")
                    break
                elif a == 2:
                    withdraw = int(input("Enter the amount to withdraw: "))
                    if withdraw > bal:
                        print("Insufficient funds.")
                    else:
                        bal -= withdraw
                        cursor.execute('UPDATE users SET balance = ? WHERE id = ?', (bal, 1))
                        conn.commit()
                        print(f"New balance: {bal}")
                    break
                elif a == 3:
                    deposit = int(input("Enter the amount to deposit: "))
                    bal += deposit
                    cursor.execute('UPDATE users SET balance = ? WHERE id = ?', (bal, 1))
                    conn.commit()
                    print(f"New balance: {bal}")
                    break
                elif a == 4:
                    new_pin = int(input("Enter new pin: "))
                    pin = new_pin
                    cursor.execute('UPDATE users SET pin = ? WHERE id = ?', (pin, 1))
                    conn.commit()
                    print("PIN changed successfully.")
                    break
            else:
                trials -= 1
                print("Incorrect pin, try again.")
                
        else:
            print("Too many incorrect attempts. Please remove your ATM card.")
            break
    else:
        print("Invalid Input")
        
conn.close()
```
